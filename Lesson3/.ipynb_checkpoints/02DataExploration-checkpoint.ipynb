```python
#Seaborn is a library for making attractive and informative statistical graphics in Python. 
#It is built on top of the matplotlib library and integrated with the PyData Stack, 
#including support for numpy and pandas data structures, and statistical routines.
```

```python
#Plotting a Heatmap with Seaborn and matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
correlations = df.corr()
sns.heatmap(data = correlations,square = True, cmap = "bwr") #Change cmap value to Blues to see changes, try ywr for error but to see other options
#For First Parameter - Rectangular dataset (2D dataset that can be coerced into an ndarray)
#For Second Parameter - If True, set the Axes aspect to “equal” so each cell will be square shaped
#For Third Parameter - Matplotlib colormap name or object, or list of colors
plt.yticks(rotation=0)
plt.xticks(rotation=90)


#Below a heatmap obtained, where, approaching red colour means maximum correlation and approaching blue means minimal correlation. Thats fir this color pattern
```
